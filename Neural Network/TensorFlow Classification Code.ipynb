```python
#Creating train and test
X = df.drop('benign_0__mal_1',axis = 1).values
y = df['benign_0__mal_1'].values
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# Part Two - Dealing with Overfitting and Evaluation

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
X_train.shape
```

```python
#CReating a Sequential Model
model = Sequential()

model.add(Dense(30,activation = 'relu'))
model.add(Dense(15,activation = 'relu'))

#This is a BINARY CLASSIFICATION problem, therefore the activation will be sigmoid
model.add(Dense(1,activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```python
#Train the model
model.fit(x = X_train, y = y_train,
         validation_data = (X_test,y_test),
         epochs = 600)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

Our model is overfitting. we can see that as the model is training, the loss is decreasing, but then there is a point when the loss is decreasing on the training data but it is increasing on the validation data. This proves that our model has overfit.

### Use 'CallBack' to prevent the model from overfitting

```python
#We will recreate the Neural Network model
model = Sequential()

model.add(Dense(30,activation = 'relu'))
model.add(Dense(15,activation = 'relu'))

#This is a BINARY CLASSIFICATION problem, therefore the activation will be sigmoid
model.add(Dense(1,activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```python
#Using 'Call Back' feature
from tensorflow.keras.callbacks import EarlyStopping
```

```python
help(EarlyStopping)
```

```python
#Creating object for 'EarlyStopping'
early_stop = EarlyStopping(monitor = 'val_loss',mode = 'min',verbose = 1,patience = 25)
```

```python
#Pass in callback while training the model
model.fit(x = X_train, y = y_train,
         validation_data = (X_test,y_test),
         epochs = 600,
         callbacks = [early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

### Another way of preventing overfitting is by adding 'DropOut' Layers in the model

```python
from tensorflow.keras.layers import Dropout
```

```python
#Pasting the model from above
model = Sequential()

model.add(Dense(30,activation = 'relu'))
model.add(Dropout(0.5))  #So we add a 'Dropout' layer after each hidden layer

model.add(Dense(15,activation = 'relu'))
model.add(Dropout(0.5))  #So we add a 'Dropout' layer after each hidden layer

#This is a BINARY CLASSIFICATION problem, therefore the activation will be sigmoid
model.add(Dense(1,activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```python
#Fitting the model
model.fit(x = X_train, y = y_train,
         validation_data = (X_test,y_test),
         epochs = 600,
         callbacks = [early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

```python
#Predicting the classification
predict_x=model.predict(X_test) 
classes_x=np.argmax(predict_x,axis=1)
```
