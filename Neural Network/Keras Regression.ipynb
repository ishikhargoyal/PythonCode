# Part Two - Data Preprocessing and Model Creation

```python
#So we now need to create test, train split and for that we will create two numpy array X and y (label)

X = df.drop('price',axis = 1).values #adding 'values' method ensures creation of numpy array instead of pd dataframe
y = df['price'].values
```

```python
X
```

```python
#Now we will do train test split
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#Normalising the scale of the data
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

## Now we will create the model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#now we will check the shape of the training data so as to make the equal number of neurons
X_train.shape
```

```python
#Creating Neural Network model
model = Sequential()

model.add(Dense(19,activation = 'relu'))
model.add(Dense(19,activation = 'relu'))
model.add(Dense(19,activation = 'relu'))
model.add(Dense(19,activation = 'relu'))

model.add(Dense(1))

model.compile(optimizer = 'adam',loss = 'mse')
```

```python
#Fitting the model
model.fit(x = X_train,
          y = y_train,
         validation_data = (X_test,y_test),
         batch_size = 128,
         epochs = 400)
```

# Part Three - Predict

```python
losses = pd.DataFrame(model.history.history)
```

```python
plt.figure(figsize = (10,18))
losses.plot()
```

```python
#now we will look at the metrics for linear regression
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
mean_absolute_error(y_test,predictions)
```

```python
#checking the explained variance score
explained_variance_score(y_test,predictions)
```

```python
plt.figure(figsize = (10,6))
plt.scatter(x=y_test,y=predictions)
```
