```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL METHOD
plt.plot(x,y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented Method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
```

# Matplotlib Part 2

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')


axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

**Figure Size (Aspect Ratio) and DPI**

```python
fig,axes =plt.subplots(nrows =2,ncols=1,figsize = (8,2))

axes[0].plot(x,y)
axes[1].plot(x,y)

plt.tight_layout()
```

```python
#HOW TO SAVE A FIGURE? use 'savefig' method
fig.savefig('my_picture.png')
```

```python
#HOW TO ADD VISUALIZATION IN OBJECT ORIENTED VISUALIZATION?

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
ax.set_title('My Title')
ax.set_xlabel('X label')
ax.set_ylabel('Y Label')
```

```python
#HOW TO ADD TWO GRAPHS ON THE SAME PLOT WITH LEGENDS DEFINED?

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#for defining the legend, we have to mention the label in the plot function before calling in the legend method
ax.plot(x,x**2,label = 'Square')
ax.plot(x,x**3,label = 'Cube')

ax.legend(loc = (0.1,0.1))
```

# Matplotlib Part 3

In this section we will be looking at how to customize the visualization from line width to line color options.

```python
#PLOT APPEARANCE - SETTING COLORS
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'purple') #You can also add RGB Hex code in the color attribute
```

```python
#HOW TO CHANGE LINE WIDTH

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'purple',linewidth = 3,linestyle = '-.') #alpha attribute is used for transparency in the line.
#linestyle is used for defining the kind of line we want in our visualization
```

```python
#HOW TO ADD A POINT MARKER

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'purple',linewidth = 3,linestyle = '-',marker = 'o',markersize = '10',
       markerfacecolor = 'yellow',markeredgewidth = '3',
       markeredgecolor = 'green') #use the marker attribute
#and also we can change the size of the marker by inputting the 'markersize' argument
```

```python
#HOW TO SET X AND Y LIMIT IN THE PLOT

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'purple',linewidth = 2, linestyle ='--')

#define the x axis limits
ax.set_xlim([0,1]) #set_xlim([lower bound,upper bound])
ax.set_ylim([0,2]) #set_xlim([lower bound,upper bound])
```
